```python
import numpy as np 
import pandas as pd 

diet_data = pd.read_csv('./foodDiet.csv')
diet_data
```

```python
from scipy.stats import f_oneway

f_static, p_value = f_oneway(diet_data['pre.weight'], diet_data['Diet'])

alpha = 0.05
print(p_value)

if p_value <= alpha:
    print('Reject hypothesis')
else:
    print('Fail to reject hypothesis')
```
